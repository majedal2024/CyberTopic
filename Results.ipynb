```python
import plotly.express as px
graph_df =resultsDS[(resultsDS["Dataset"] == "TopicCybertweets") & (resultsDS["labellingType"] == "labelsCyberAttack")]
#graph_df =resultsDS[(resultsDS["Dataset"] == "TopicCybertweets") & (resultsDS["labellingType"] == "labelszeroday")]
#graph_df =resultsDS[(resultsDS["Dataset"] == "Twitter_DS") & (resultsDS["labellingType"] == "labelsCyberAttack")]
#graph_df =resultsDS[(resultsDS["Dataset"] == "Twitter_DS") & (resultsDS["labellingType"] == "labelszeroday")]
fig = px.line(graph_df, x="NumberOfTopic", y="F1_ScoreScore", color='Model',markers=True)
fig.update_layout(
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)',
        ),
    ),
    yaxis=dict(
        showgrid=False,
        zeroline=False,
        showline=False,
        showticklabels=True,
    ),
    autosize=False,
    margin=dict(
        autoexpand=True,
        l=100,
        r=20,
        t=110,
    ),
    showlegend=True,
    plot_bgcolor='white',
    title='F1 Results for for X\'s Cybersecurity Dataset 2022 - All Cyberattacks',
    xaxis_title='Number Of Topic (K)',
    yaxis_title='F1 Score'
)
fig.show()
```
