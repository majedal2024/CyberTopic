```python
from openai import OpenAI
openai_key = ''
openai_org= ''
client = OpenAI(api_key=openai_key,organization=openai_org)


# Function to classify Urgency
def classify_Urgency(tweet, openai_key):
    client.api_key = openai_key

    response = client.completions.create(
        model="text-davinci-002",
        prompt= f""" Classify the following tweets into four categories based on their cybersecurity relevance:
        '1': Tweets not related to cybersecurity.
        '2': Broad discussions about cybersecurity without specific threats or incidents.
        '3': Mentions of vulnerabilities or bugs, without referencing active cyberattacks or exploits.
        '4': Specific, ongoing cyber threats like active attacks, data breaches, or malware exploitation.
        '5': Mentions of fix or update of cyberattacks or exploits.
      Categorize each tweet as '1', '2', '3', '4' or '5' as the above categiries.\n\n{tweet}\n\nLabel: """
    )
    # 1: Unrelated
    # 2: General
    # 3: Warning
    # 4: Critical
    nLabel = response.choices[0].text.strip().lower()
    return nLabel

# Example usage
tweets = df_Twitter['Tweet']
ds_label = []
for tweet in tweets:
  u_Label = classify_Urgency(tweet, openai_key)
  ds_label.append(u_Label)
    #print(f"Tweet: '{tweet}' - Chat GPT Label: {u_Label}")
df_Twitter['label']= ds_label
df_Twitter.to_csv('/content/drive/My Drive/PhD/Final/Twitter_DS.csv')
```
